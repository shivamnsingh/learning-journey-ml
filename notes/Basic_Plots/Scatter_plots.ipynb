```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Superstore.csv', encoding='latin1')
```

```python
df.columns
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Discount'], df['Profit'], alpha=0.5,color='seagreen')
plt.title('Discount vs Profit Scatter Plot')
plt.xlabel('Discount')
plt.ylabel('Profit')
```
