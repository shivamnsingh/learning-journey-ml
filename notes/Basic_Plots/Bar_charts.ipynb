```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Superstore.csv', encoding='latin 1')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.set_index('Order Date').resample('M')['Sales'].sum().reset_index()

plt.figure(figsize=(10,6))
plt.bar(monthly_sales['Order Date'], monthly_sales['Sales'], color='royalblue', width=20)

plt.title('Monthly Sales (Vertical Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.set_index('Order Date').resample('M')['Sales'].sum().reset_index()
months = monthly_sales['Order Date'].dt.strftime('%Y-%m')

plt.figure(figsize=(10,6))
plt.barh(months, monthly_sales['Sales'], color='royalblue')

plt.title('Monthly Sales (Horizontal Bar Chart)')
plt.xlabel('Total Sales')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
```
