```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Superstore.csv', encoding='latin1') 
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
df.describe()
```

```python
monthly_sales = df.set_index('Order Date').resample('M')['Sales'].sum().reset_index()
```

```python
plt.figure(figsize=(10,6))
plt.step(monthly_sales['Order Date'], monthly_sales['Sales'], where='mid', marker='o', color='RoyalBlue')
plt.title('Monthly Sales Step Plot')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
